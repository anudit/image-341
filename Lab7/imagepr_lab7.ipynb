```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data 
from PIL import Image, ImageOps, ImageEnhance 
%matplotlib inline
```

### Q1

```
im = Image.open("parrot.png")
im_invert = ImageOps.invert(im)

fig, axs = plt.subplots(1, 2, figsize=(20,20))
axs[0].imshow(im)
axs[0].set_title('original')

axs[1].imshow(im_invert)
axs[1].set_title('inverted')
plt.show()
```

### Q2

```
img = cv2.imread('parrot.png') 

c = 255/(np.log(1 + np.max(img))) 
log_transformed = c * np.log(1 + img) 
log_transformed = np.array(log_transformed, dtype = np.uint8) 

fig, axs = plt.subplots(1, 2, figsize=(20,20))
axs[0].imshow(img)
axs[0].set_title('original')

axs[1].imshow(log_transformed)
axs[1].set_title('log transformed')
plt.show()
```

### Q3

```
img = cv2.imread('parrot.png')

img_r, img_g, img_b = cv2.split(img)

fig, axs = plt.subplots(1, 4, figsize=(20, 4))

axs[0].imshow(img)
axs[0].set_title('original image')

axs[1].hist(img_r.ravel(), bins=256)
axs[1].set_title('R')

axs[2].hist(img_g.ravel(), bins=256)
axs[2].set_title('G')

axs[3].hist(img_b.ravel(), bins=256)
axs[3].set_title('B')

plt.show()
```

```
from skimage import exposure

img = cv2.imread('parrot.png')
img_eq = exposure.equalize_hist(img)
img_r, img_g, img_b = cv2.split(img_eq)

fig, axs = plt.subplots(1, 4, figsize=(20, 4))

axs[0].imshow(img_eq)
axs[0].set_title('Equalized image')

axs[1].hist(img_r.ravel(), bins=256)
axs[1].set_title('R')

axs[2].hist(img_g.ravel(), bins=256)
axs[2].set_title('G')

axs[3].hist(img_b.ravel(), bins=256)
axs[3].set_title('B')

plt.show()

```

### Q4

```
from skimage import exposure

img = cv2.imread('parrot.png')
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))

img_r, img_g, img_b = cv2.split(img_rescale)

fig, axs = plt.subplots(1, 4, figsize=(20, 4))

axs[0].imshow(img_rescale)
axs[0].set_title('Contrast Stretched Image')

axs[1].hist(img_r.ravel(), bins=256)
axs[1].set_title('R')

axs[2].hist(img_g.ravel(), bins=256)
axs[2].set_title('G')

axs[3].hist(img_b.ravel(), bins=256)
axs[3].set_title('B')

plt.show()
```
