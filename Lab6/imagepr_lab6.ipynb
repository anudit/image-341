## Q1

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data 
from skimage.filters import roberts, sobel, scharr, prewitt
from skimage.feature import canny
%matplotlib inline
```

```
image = data.camera()

fig, axs = plt.subplots(1, 5, figsize=(20,20))
axs[0].imshow(image, cmap='gray')
axs[0].set_title('original')

axs[1].imshow(roberts(image), cmap='gray')
axs[1].set_title('roberts')

axs[2].imshow(sobel(image), cmap='gray')
axs[2].set_title('sobel')

axs[3].imshow(canny(image), cmap='gray')
axs[3].set_title('canny')

axs[4].imshow(prewitt(image), cmap='gray')
axs[4].set_title('prewitt')

plt.show()
```

## Q2

```
from scipy import misc, signal
from scipy import ndimage, misc
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu, laplace, gaussian

original = cv2.imread('img1.png', 1)
im = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)

laplace_kernel = np.array(im)
im_edges = laplace(im, ksize=3)
log = laplace(gaussian(im, sigma=0.8),ksize=3)

fig, axs = plt.subplots(1, 3, figsize=(15,15))
axs[0].imshow(im)
axs[0].set_title('original')

axs[1].imshow(im_edges)
axs[1].set_title('Laplace')

axs[2].imshow(log)
axs[2].set_title('LoG')

plt.show()
```

## Q3

```
from skimage.filters import laplace, gaussian, sobel

original = cv2.imread('img2.png', 1)
im = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)

lap = laplace(im, ksize=10)
im2 = im+lap
sob = sobel(im2)
sm = gaussian(sob, sigma=5)
mas = sobel(im2, mask = sm)
fin = im + mas

fig, axs = plt.subplots(1, 7, figsize=(22,22))
axs[0].imshow(im, cmap='gray')
axs[0].set_title('A')

axs[1].imshow(lap, cmap='gray')
axs[1].set_title('B')

axs[2].imshow(im2, cmap='gray')
axs[2].set_title('C')

axs[3].imshow(sob, cmap='gray')
axs[3].set_title('D')

axs[4].imshow(sm, cmap='gray')
axs[4].set_title('E')

axs[5].imshow(mas, cmap='gray')
axs[5].set_title('F')

axs[6].imshow(fin, cmap='gray')
axs[6].set_title('G')

plt.show()
```
